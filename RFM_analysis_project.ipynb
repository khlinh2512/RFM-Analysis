```python
import numpy as np
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read excel file with sheet name
dict_df = pd.read_excel('Dataset.xlsx', sheet_name= ["Orders", "Return","Segmentation"])
```

```python
# Get DataFrame from Dict
orders = dict_df.get('Orders')
returns = dict_df.get('Return')
segments = dict_df.get('Segmentation')
```

```python
print(orders.info())
print(returns.info())
print(segments.info())
```

```python
orders.head(10)
```

```python
#Remove the returned orders from the orders table.

df_not_return = orders[~orders['Order ID'].isin(returns[returns['Returned'] == 'Yes']['Order ID'])]

df_not_return.shape
```

```python
# Remove duplicate rows from the orders DataFrame
orders.drop_duplicates(inplace=True)
```

```python
#Calculating RFM of each customer

recent_date = pd.to_datetime("2017-12-31")

#create new dataframe concluding R, F, M group by each customer ID
df_rfm = df_not_return.groupby('Customer ID', as_index=False).\
            agg({'Order Date' : lambda date: (recent_date - date.max()).days,
                 'Order ID' : 'nunique',
                 'Sales' : 'sum'})

# Change the name of columns from dataframe df_rfm
df_rfm.columns = ['Customer ID','recency', 'frequency', 'monetary']
 
df_rfm
```

```python
score_recency = [5,4,3,2,1]
score = [1,2,3,4,5]

#Create a new column to store the quintile score of recency
df_rfm['recency_score'] = pd.qcut(df_rfm.recency, q=5, labels = score_recency)

#Create a new column to store the quintile score of frequency
df_rfm['frequency_score'] = pd.qcut(df_rfm.frequency, q=5, labels = score)

#Create a new column to store the quintile score of monetary
df_rfm['monetary_score'] = pd.qcut(df_rfm.monetary, q=5, labels = score)

df_rfm
```

```python
#Concatnation r,f,m into rfm column as string type
df_rfm['rfm'] = df_rfm['recency_score'].astype(str) +\
                df_rfm['frequency_score'].astype(str) + \
                df_rfm['monetary_score'].astype(str)

df_rfm['rfm'] = df_rfm['rfm'].astype(int)

df_rfm
```

```python
rfm_segment_map = {}

for row in segments.itertuples(index=False):
    segment = row.Segment
    rfm_score_ranges = [
        int(score.strip())
        for score in row._1.split(',')
        ]
    for rfm_score_range in rfm_score_ranges:
        rfm_segment_map[rfm_score_range] = segment

rfm_segment_map
```

```python
segment_new = pd.DataFrame.from_dict(rfm_segment_map, orient="index", columns=["Segment"])

segment_new = segment_new.reset_index() # reset index to move dictionary keys to a column
segment_new = segment_new.rename(columns={'index': 'rfm'}) # rename column to 'rfm'

segment_new
```

```python
#Join df_rfm and segment_new
rfm_merge = df_rfm.merge(segment_new, on = 'rfm', how = 'inner')

rfm_merge = rfm_merge.sort_values('Customer ID')
rfm_merge
```

```python
colname = ['recency', 'frequency', 'monetary']

for i in colname:
    fig, ax = plt.subplots(figsize = (9,3))
    sns.distplot(rfm_merge[i])
    ax.set_title(f"Distribution of {i}")
    plt.show
```

```python
colname1 = ['recency', 'frequency', 'monetary','recency']

for i in range(3):
    fig, ax = plt.subplots(figsize = (9,3))
    sns.scatterplot(x= colname1[i], y=colname1[i+1], 
                data=rfm_merge)
    ax.set_title(f"{colname1[i]} vs {colname1[i+1]}")
    plt.show()
    
```

```python
rfm1 = rfm_merge
rfm1['monetary1'] = rfm1['monetary']
aggta = rfm_merge.groupby('Segment', as_index=False).agg({
                'Customer ID': 'count',
                'recency':'mean',
                'frequency': 'mean',
                'monetary': 'mean',
                'monetary1': 'sum'
})
aggta.columns = ['segment',"customer_num", "r_avg", "f_avg", "m_avg", "m_sum"]
aggta = aggta.sort_values('m_sum', ascending = False)
aggta
```

```python
#số lượng khách hàng phân nhóm theo phân khúc
segment_order = ['Champions', 'Loyal', 'Potential Loyalist', 'New Customers', 'Promising', 
                 'Need Attention', 'About To Sleep', 'At Risk', 'Cannot Lose Them', 
                 'Hibernating customers','Lost customers']

cols = ['customer_num', 'r_avg', 'f_avg','m_avg']
col_interpret = ['Customer Count', 'mean of Recency', 'mean of Frequency', 'mean of Total sales']

for i in range(4):
    fig, ax = plt.subplots(figsize = (10,3))
    
    sns.barplot(x='segment', y=cols[i], data=aggta, order = segment_order)
    
    plt.title(f'RFM Segments of {col_interpret[i]}')
    plt.xticks(rotation = 40)
    plt.show()
```

## Mục đích: Xác định các nhóm khách hàng cần tập trung để đưa ra những chiến lược phù hợp (theo 80/20 rule)

1. Các nhóm nhiều khách hàng nhất: **New Customer, At risk**

 (trong khi hai phân khúc chúng ta quan tâm nhất được dự báo là mang lại doanh thu và lợi nhuận lớn nhất là Champions và Loyal Customers thì chỉ chiếm phần nhỏ khách hàng. Điều này phù hợp với "the 80/20 rule" tức “80% doanh thu doanh nghiệp của bạn đến từ 20% khách hàng”)
 
 
2. Các nhóm mua gần đây nhất: **Champions, Potential Loyalist, Promising, New Customer, Need Attention and Loyal**


3. Các nhóm tần suất mua nhiều: **Champion, Loyal, Cannot Lose Them, At risk**
 
 Để gia tăng tỉ lệ retention, chú ý hơn và các nhóm Potential Loyalist và About to Sleep
 
 
4. Các nhóm chi nhiều tiền: **Champion, Loyal, Cannot Lose Them, At risk**
  
  Nhóm Cannot Lose Them, At risk chi nhiều, và đã từng thường xuyên trong quá khứ, tuy nhiên đã lâu ko mua -> Cần có thêm những giải pháp để giữ chân những khách hàng này



```python
# Tính tổng m_sum của tất cả các nhóm khách hàng
total_sales = aggta['m_sum'].sum()

# Tạo list giá trị cho pie chart (tỉ lệ của mỗi segment)
pie_values = [aggta.loc[i,'m_sum']/total_sales for i in aggta.index]

# Vẽ pie chart
plt.figure(figsize=(5,5))
plt.pie(pie_values, labels=aggta['segment'], autopct='%1.1f%%', startangle=0)
plt.axis('equal')
plt.title('Segmentation by Total Sales - %')
plt.show()
```

```python
# Vẽ pie chart the number of customer
plt.figure(figsize=(5,5))
plt.pie(aggta['customer_num'], labels=aggta['segment'], autopct='%1.1f%%', startangle=0)
plt.axis('equal')
plt.title('Segmentation by Customer number - %')
plt.show()
```

## Summary: 

* Với tiêu chí:
    * A cost of business is 5 times more to attract a new customer than to retain the existing ones.
    * 20% of customer contributing to 80% of total revenues
    * Do đó: Cần đặc biệt chú ý vào những segment khác hàng mang lại doanh thu cao và thường xuyên

1. **Loyal, Champions (VIP customer)** *- thường xuyên đóng góp lớn trong total sales*

* Actionable tip:

    * Reward them. Can be early adopters for new products. Will promote your brand
    * Upsell higher value products. Ask for reviews. Engage them.
    * Đặc biệt phải dành sự ưu tiên, tránh gây ra sự mất lòng đối với những VIP customer này
    * Nhanh chóng triển khai các chương trình dành cho hai phân khúc khách hàng VIP với kỳ vọng mang lại doanh số giải ngân và lợi nhuận tốt nhất cho công ty
    

2. **Potential Loyalist** - *khách hàng tiềm năng trở thành khách hàng thân thiết mang lại lợi nhuận lâu dài cho công ty.*

* Actionable tip:

     * Offer membership / loyalty program, recommend other products.
     * Có chính sách hoặc xây dựng hệ thống sản phẩm hiệu quả, linh hoạt hơn nhằm khai thác lượng lớn khách hàng
     
    
3. **At Risk và Cannot Lose Them** *- 2 segment với khối lượng mua lớn, chỉ là đã lâu ko mua*

* Actionable tip:

    * Send personalized emails to reconnect, offer renewals, provide helpful resources.
    * Win them back via renewals or newer products, don’t lose them to competition, talk to them.
    * Cẩn trọng trong việc điều chỉnh các chính sách cũng như biện pháp tức thời vì tuy nhóm khách hàng này có tỷ lệ rủi ro cao nhưng đang mang lại doanh số giải ngân cũng rất cao.
